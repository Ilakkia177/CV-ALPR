```python
from car_detection import detect_cars
from plate_detection import detect_plates
from plate_segmentation import segment_characters
from template_matching import load_templates, recognize_plate
from vizualize_pipeline import draw_annotations      # ‚úÖ fixed typo: ‚Äúvizualize_pipeline‚Äù ‚Üí ‚Äúvisualize_pipeline‚Äù
from utils.interpolation import interpolate_boxes  # ‚úÖ match actual filename
import pandas as pd, os, cv2

# ------------------ PATH SETUP ------------------
os.chdir("/Users/ilakkiat/Documents/MIT/SEM-V/CV-LAB/CV-PROJECT/CODE")
print("Current working directory:", os.getcwd())

VIDEO_PATH = "data/cropped_10.mp4"
CARS_DIR = "outputs/cars"
PLATES_DIR = "outputs/plates"
TEMPLATES_DIR = "templates/char_templates_hybrid"
RESULTS_CSV = "outputs/results.csv"
INTERP_CSV = "outputs/interpolated_results.csv"
FINAL_VIDEO = "outputs/stabilized_output.mp4"

# ------------------ 1Ô∏è‚É£ Detect Cars ------------------
detect_cars(VIDEO_PATH, CARS_DIR)

# ------------------ 2Ô∏è‚É£ Detect Plates ------------------
detect_plates(CARS_DIR, PLATES_DIR)

# ------------------ 3Ô∏è‚É£ Recognize Plate Text ------------------
templates = load_templates(TEMPLATES_DIR)

data = []
for file in os.listdir(PLATES_DIR):
    if not file.lower().endswith((".jpg", ".png")):
        continue
    img_path = os.path.join(PLATES_DIR, file)
    img = cv2.imread(img_path)
    if img is None:
        print(f"‚ö†Ô∏è Skipping unreadable image: {file}")
        continue
    chars = segment_characters(img)
    if not chars:
        print(f"‚ùå No characters found in: {file}")
        continue
    text = recognize_plate(chars, templates)
    print(f"‚úÖ {file}: {text}")
    data.append([file, text])

df = pd.DataFrame(data, columns=["plate_image", "plate_text"])
df.to_csv(RESULTS_CSV, index=False)
print(f"‚úÖ Results saved to {RESULTS_CSV}")

# ------------------ 4Ô∏è‚É£ Postprocess: Interpolation ------------------
interpolated = interpolate_boxes(RESULTS_CSV)
interpolated.to_csv(INTERP_CSV, index=False)
print(f"‚úÖ Interpolated CSV saved to {INTERP_CSV}")

# ------------------ 5Ô∏è‚É£ Overlay on Video ------------------
draw_annotations(VIDEO_PATH, INTERP_CSV, FINAL_VIDEO)
print(f"üé• Final annotated video saved as {FINAL_VIDEO}")
```
